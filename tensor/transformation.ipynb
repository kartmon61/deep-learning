```python
import torch
import torchvision
```

```python
from PIL import Image

img = Image.open('sample_images/cat_img.jpeg')
```

```python
img
```

```python
gaussian_blur = torchvision.transforms.GaussianBlur(kernel_size=5, sigma=(6.0, 6.0))
```

```python
gaussian_blur(img)
```

```python
random_affine = torchvision.transforms.RandomAffine(degrees=30, translate=(0.1, 0.1), scale=(0.8, 1.2), shear=15)
```

```python
random_affine(img)
```

```python
combined_transform = torchvision.transforms.Compose([gaussian_blur, random_affine])
```

```python
combined_transform(img)
```

```python
from torchvision.transforms import Lambda
```

```python
target_transform = Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(dim=0, index=torch.tensor(y), value=1))
```

```python
target_transform(0)
```

```python
target_transform(1)
```
